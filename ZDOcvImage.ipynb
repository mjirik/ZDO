## Color to gray

```python
import skimage.color
img = skimage.color.rgb2gray(im)
print(img.shape)
plt.imshow(img)
plt.colorbar()
```

# Plot

```python
plt.imshow(img[50:350, 150:350, 0], cmap='gray')
```
